```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
```

```python
G = nx.DiGraph()

G.add_edge('A', 'B')
G.add_edge('A', 'D')
G.add_edge('B', 'C')
# G.add_edge('D', 'E')
G.add_edge('E', 'C')
G.add_edge('D', 'B')
G.add_edge('B', 'E')

pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos, with_labels=True, node_size=2000, node_color=['tab:orange', 'lightgrey', 'lightgrey', 'tab:green', 'lightgrey'], font_size=20, arrowsize=25)
```

```python
F = nx.Graph()

for i in range(5):
    F.add_node(f"F[{i}]")

    if i > 1:
        F.add_edge(f"F[{i}]", f"F[{i-1}]")
    if i > 2:
        F.add_edge(f"F[{i}]", f"F[{i-2}]")


nx.draw(F, pos={f"F[{i}]": (0, i) for i in range(5)}, with_labels=True, node_size=2000, node_color='lightgrey', font_size=20)
```
